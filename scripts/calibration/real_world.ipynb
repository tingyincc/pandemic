```python
import pandas as pd

df = pd.read_csv("owid-covid-data.csv")
```

```python
df.head()
```

```python
df_italy = df[df["location"]=="Italy"]
```

```python
df_italy["date"]
```

```python
df_italy = df_italy[pd.to_datetime(df_italy["date"]) < pd.to_datetime("2020-05-31")]
```

```python
df_italy
```

```python
import numpy as np
df_italy["date"] = np.datetime64(df_italy["date"])
```

```python
df_italy = df_italy.sort_values('date', ascending=True)
```

```python
df_italy["date"]
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

x=df_italy['date']
y=df_italy['new_cases']
fig, ax = plt.subplots(figsize=(30,10))
ax.plot(x, y)
start, end = ax.get_xlim()
print(start, end)
ax.xaxis.set_ticks(np.arange(start, end, 5))
plt.xticks(rotation=45)
plt.show()
```

```python
x=df_italy['date']
fig, ax = plt.subplots(figsize=(30,10))
# ax.plot(x, df_italy['new_cases'], label = "new case")
ax.plot(x, df_italy['new_cases_smoothed'], label = "new case")
# ax.plot(x, df_italy['new_cases'], label = "total case")
ax.plot(x, df_italy['new_deaths_smoothed'], label = "dead case", color = "darkgrey")
# ax.plot(x, df_italy['hosp_patients'], label = "hospital patients")
ax.plot(x, df_italy['icu_patients'], label = "icu patients", color = "red")
start, end = ax.get_xlim()
print(start, end)
ax.xaxis.set_ticks(np.arange(start, end, 7))
ax.legend()
plt.xticks(rotation=45)
plt.show()
```
