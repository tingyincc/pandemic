```python
def draw_two_sim_compare(real_df, sim1_df, sim2_df, sim_factor=1, title1="", title2 =""):
    x=real_df['date']
    fig, ax = plt.subplots(figsize=(30,10))
    # ax.plot(x, df_italy['new_cases'], label = "new case")
    # ax.plot(x, sim1_df['Infected']*sim_factor, label = "simulated infected", color = "orange",marker='+',linestyle='dotted')
    #ax.plot(x, sim1_df['Dead']*sim_factor, label = "simulated dead"+title1, color = "grey",marker='.',linestyle='dotted')
    ax.plot(x, sim1_df['Critical']*sim_factor, label = "simulated critical"+title1,color = "red",marker='.',linestyle='dotted')
    # ax.plot(x, sim2_df['Infected']*sim_factor, label = "simulated infected", color = "orange",marker='+',linestyle='dashed')
    #ax.plot(x, sim2_df['Dead']*sim_factor, label = "simulated dead"+title2, color = "grey",marker='+',linestyle='dashed')
    ax.plot(x, sim2_df['Critical']*sim_factor, label = "simulated critical"+title2,color = "red",marker='+',linestyle='dashed')
    start, end = ax.get_xlim()
    print(start, end)
    ax.xaxis.set_ticks(np.arange(start, end, 7))
    ax.legend()
    plt.xticks(rotation=45)
    plt.show()
```

```python
def draw_real_sim_compare(real_df, sim_df, sim_factor=6, offset=0):
    x=real_df['date']
    fig, ax = plt.subplots(figsize=(20,10))
    # ax.plot(x, df_italy['new_cases'], label = "new case")
    # ax.plot(x, real_df['new_cases_smoothed'], label = "new case",color = "orange", marker='o')
    #ax.plot(x, df_italy['new_cases'], label = "total case")
    #ax.plot(x, real_df['total_deaths'], label = "dead case", color = "grey", marker='o')
    #ax.plot(x, df_italy['hosp_patients'], label = "hospital patients")
    ax.plot(x, real_df['icu_patients'][offset:].append( real_df['icu_patients'][:offset]), label = "icu patients", color = "red", marker='o')
    #ax.plot(x, sim_df['Infected']*sim_factor, label = "simulated infected", color = "orange",marker='+',linestyle='dashed')
    #ax.plot(x, sim_df['Dead']*sim_factor, label = "simulated dead", color = "grey",marker='+',linestyle='dashed')
    ax.plot(x, sim_df['Critical']*sim_factor, label = "simulated critical",color = "red",marker='+',linestyle='dashed')
    start, end = ax.get_xlim()
    print(start, end)
    ax.xaxis.set_ticks(np.arange(start, end, 7))
    ax.legend()
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 100 linearly spaced numbers
x = np.linspace(0,120,121)

# the function, which is y = e^x here

y_poly = 1- 0.0001 * pow( x, 2)
y_poly[y_poly < 0] = 0

y_exp = (1.05 ** x) /100 
y_exp[y_exp>1] = 1

y_custom = []
for day, num in enumerate(x):
    if day < 25:
        y_custom.append(0.7 + 0.011*day)
    else:
        y_custom.append( 0.964 - 0.0001 * pow( day-25, 2))

y_custom = np.array(y_custom)

# setting the axes at the centre
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
# ax.spines['left'].set_position('center')
# ax.spines['bottom'].set_position('zero')
# ax.spines['right'].set_color('none')
# ax.spines['top'].set_color('none')
# ax.xaxis.set_ticks_position('bottom')
# ax.yaxis.set_ticks_position('left')

# plot the function
ax.plot(x,y_exp,label='exponential growth')
ax.plot(x,y_poly, label='polynomial decay')
ax.plot(x,y_custom, label='custom')
plt.title("Compliance Probability Function")
plt.legend()

# show the plot
plt.show()
```
